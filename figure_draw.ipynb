```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
data = pd.read_csv('data/timing_results.csv')  # Replace with the actual path to your CSV file

# Convert the 'Number of switches' column to integer, if not already
data['Calculation Time (ms)'] = data['Calculation Time (ms)'] /1000

# Split the data by graph type
erdos_renyi_data = data[data['Type of graph'] == 'Erdos-Renyi']
waxman_data = data[data['Type of graph'] == 'Waxman']

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(erdos_renyi_data['Number of switches'], erdos_renyi_data['Calculation Time (ms)'], marker='x', color='red', label='Erdos-Renyi graph')
plt.plot(waxman_data['Number of switches'], waxman_data['Calculation Time (ms)'], marker='x', color='green', label='Waxman graph')

# Adding labels and title
plt.xlabel('The number of switches')
plt.ylabel('Compute time (s)')
plt.title('The total computing time vs. the number of switches in the network with 20000 active flows')
switches = data['Number of switches'].unique()
plt.xticks(switches[switches >= 200], rotation=45) 
plt.legend()

# Show the plot
```
